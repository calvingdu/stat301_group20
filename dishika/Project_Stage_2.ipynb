# **Section 4: Method and Plan**

- **Proposed Method:**
  - I will use an Additive Multivariable Logistic Regression model with Gallstone Status (0 = gallstones present, 1 = no gallstones) as the binary response and all 38 clinical features as explanatory variables, including demographic variables, bioimpedance measures, and laboratory features.
  - I will first convert all categorical variables to factors and then fit the model using glm.
- **Why this method is appropriate:** Logistic regression is appropriate because the outcome variable is binary and my aim is to quantify how each clinical factor is associated with the odds of gallstone disease while keeping all other variables constant. This model directly supports my goal of inference.
- **Key assumptions:**
    - The model assumes that each patient's gallstone status is independent of other patient's outcomes.
    - There is a linear relationship between each continuous predictor and the log odds of gallstone disease.
    - The model assumes there is no multicollinearity.
    - The sample size is large enough for the model to be valid.
- **Potential limitations:** Including all 38 features in one additive logistic regression model can lead to overfitting and unstable coefficient estimates. Moreover, the additive model may not capture important interactions between variables.

# **Section 5:Computational Code and Output**

```r
library(readxl)     
library(dplyr)      
library(janitor)    
library(ggplot2)   
library(scales)    
library(broom)

data_raw <- read_excel("gallstone.xlsx")

gallstone <- data_raw |>
  janitor::clean_names()
```

```r
# Converting categorical variables to factors

gallstone_model <- gallstone |>
  mutate(
    gallstone_status = factor(
      gallstone_status,
      levels = c(1, 0),
      labels = c("NoGallstone", "Gallstone")
    ),
    gallstone_status = relevel(gallstone_status, ref = "NoGallstone"),

    gender = factor(
      gender,
      levels = c(0, 1),
      labels = c("Male", "Female")
    ),

    comorbidity = factor(
      comorbidity,
      levels = c(0, 1, 2, 3),
      labels = c("0", "1", "2", "3")
    ),

    coronary_artery_disease_cad = factor(
      coronary_artery_disease_cad,
      levels = c(0, 1),
      labels = c("NoCAD", "CAD")
    ),

    hypothyroidism = factor(
      hypothyroidism,
      levels = c(0, 1),
      labels = c("NoHypo", "Hypo")
    ),

    hyperlipidemia = factor(
      hyperlipidemia,
      levels = c(0, 1),
      labels = c("NoHyperlipid", "Hyperlipid")
    ),

    diabetes_mellitus_dm = factor(
      diabetes_mellitus_dm,
      levels = c(0, 1),
      labels = c("NoDM", "DM")
    ),

    hepatic_fat_accumulation_hfa = factor(
      hepatic_fat_accumulation_hfa,
      levels = c(0, 1, 2, 3, 4),
      labels = c("NoFat", "Grade1", "Grade2", "Grade3", "Grade4"),
      ordered = TRUE
    )
  )
head(gallstone_model)
```

```r
fit_model <- glm(
  gallstone_status ~ .,
  data = gallstone_model,
  family = binomial
)
```

```r
model_table <- fit_model |>
  tidy(conf.int = TRUE, exponentiate = TRUE)
model_table
```

**Interpretation:**

- Using a 5% significance level, the multivariable logistic regression indicates that age, gender female, coronary artery disease(CAD), diabetes mellitus(DM), bone mass, HFA, AST, CRP, hemoglobin, and vitamin D are significantly associated with having a gallstone disease (not odds since we are doing exponentiate = TRUE in the code) because their p-value \< 0.05. Most remaining predictors have p-values above 0.05, so there is little evidence that they are associated with gallstone status.
- The variables which have odds ratios greater than 1 suggest that they are associated with higher odds of gallstone disease, keeping other variables constant, and vice-versa.
- This method only gives few variables out of 38 that are statistically significant to the response, possibly excluding other important variables. To improve the analysis, I could use LASSO logistic regression to identify a smaller subset of important predictors and then fit an additive multivariable logistic regression model only using the selected variables.
